```python
import pandas as pd
import numpy as np
```

```python
#地点の区切り位置を見つける関数
def EndDetecter(data):
    return data[data[0].isnull()].index.values
```

```python
#終了位置のインデックスを見つける
def EndIndex(start,stops):
    for end in stops:
        if start< end:
            return end
```

```python
#IDからひとまとまりの時系列データを取り出す関数
def RetrieveOneSpot(data,ID):# IDは文字列で与える
    start = data[data[0] ==ID].index.values[0]
    ends = EndDetecter(data)
    end = EndIndex(start, ends)
    return data[start:end]
```
